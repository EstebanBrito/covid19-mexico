```python
# Función para dibujar gráficos de prueba
def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2, figsize=(11, 4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
```

```python
# Se puede usar plt.style.use('default') también
with plt.style.context('default'):
    hist_and_lines()
```

```python
# 5.38 es una página web con un estilo característico
plt.style.use('fivethirtyeight')
hist_and_lines()
```

```python
# Similar a ggplot, una librería de R
plt.style.use('ggplot')
hist_and_lines()
```

### Colores, tipos de línea y tipos de marcadores ###

```python
plt.figure()

# Color y tipo de línea pueden ser expresados diferentemente
plt.plot(x, x, color='red', linestyle='solid')
plt.plot(x, x + 1, color='b', linestyle='dashed') # blue (rgbcmyk)
plt.plot(x, x + 2, color='g', linestyle=':') # dotted (-, --, :, -.)
plt.plot(x, x + 3, '-.y') # yellow y dashdot
```

```python
x = np.linspace(0, 10, 20)
y = np.sin(x)

# Tambien se pueden agregar tipos de marcadores
plt.plot(x, y, '-^k') # Varios tipos: (o . ^ s d p x +)
```

```python
# Más personalización para los tipos de marcadores
plt.plot(x, y, '-p', color='gray', linewidth=4,
         markersize=15, markerfacecolor='white',
         markeredgecolor='gray', markeredgewidth=2)
```

Los tipos de marcadores son muy útiles en gráficas de dispersión, que veremos más adelante.

### Ejes ###

```python
plt.plot(x, np.sin(x))

# Puedes establecer límites
plt.xlim(-2, 12)
plt.ylim(-2, 2)
```

```python
plt.plot(x, np.sin(x))

# O elegir entre estilos predeterminados
plt.axis('tight') # Menos espacio entre elementos
# plt.axis('equal') # Ejes de la misma escala
```

### Leyendas y Títulos ###

```python
plt.plot(x, np.sin(x), '-r', label='sen(x)')
plt.plot(x, np.cos(x), '--b', label='cos(x)')
plt.title('Titulo de la gráfica')
plt.xlabel('x')
plt.ylabel('sen(x)')
plt.legend()

plt.axis('equal')
```

```python
# En POO se hace de la sig. manera
ax = plt.axes()
ax.plot(x, np.sin (x), label='sen(x)')
ax.set(xlim=(0, 10), ylim=(-2, 2),
      xlabel='x', ylabel='sen(x)',
      title='Gráfica')
ax.legend()
```

```python
# Creamos una función para guardar cualquier gráfica
IMGS_FOLDER = os.path.join('..', 'imgs')
if not os.path.exists(IMGS_FOLDER): os.mkdir(IMGS_FOLDER)
    
def save_fig(figure, name):
    IMG_URL = os.path.join(IMGS_FOLDER, f'{name}.png')
    figure.savefig(IMG_URL)
```

## Histogramas (Gráficos de barras) ##

```python
import pandas as pd
CLEAN_DATA_FOLDER = os.path.join('..', 'datasets', 'clean_data')
```

```python
!dir ..\datasets\clean_data
```

#### ¿Cuántos hombres y mujeres se infectaron o fallecieron? ####

```python
covid_por_genero = pd.read_csv(CLEAN_DATA_FOLDER + '\covid_por_genero.csv')
covid_por_genero
```

```python
ax = covid_por_genero.plot.bar(x='GENERO', rot=0)
ax.legend(loc='upper left')
```

```python
# Un detalle. Los AxesSubplot tiene un Figure dentro
save_fig(ax.figure, 'covid_por_genero')
```

#### ¿Qué estados han sufrido más por el COVID? ####

```python
covid_por_estado = pd.read_csv(CLEAN_DATA_FOLDER + '\covid_por_estado.csv')
covid_por_estado
```

```python
ax = covid_por_estado[::-1].plot.barh(x='ESTADOS', figsize=(10, 15), grid=True)
```

```python
save_fig(ax.figure, 'covid_por_estados')
```
